#Statistics Basics Assignment Answers

###1. What is statistics, and why is it important?

-> Statistics is a mathematical science including methods of collecting,organising and analysing data in such a way that meaningful conclusion can be drawn from them.

It is important as it helps in decision-making, understanding trends, and making predictions.

###2. What are the two main types of statistics?
-> The two main types of statistics are:

- Descriptive Statistics

- Inferential Statistics

###3. What are descriptive statistics?

-> Descriptive statistics summarize and organize data using measures like mean, median, mode, range, and standard deviation.

###4. What is inferential statistics?

-> Inferential statistics use a sample of data to make generalizations or predictions about a population.

###5. What is sampling in statistics?
-> Sampling is the process of selecting a subset (sample) from a population to analyze and make inferences about the entire population.

###6. What are the different types of sampling methods?

-> The different types of sampling methods are:

- Random Sampling

- Stratified Sampling

- Systematic Sampling

- Cluster Sampling

- Convenience Sampling

###7. What is the difference between random and non-random sampling?

-> The difference between random and non-random sampling are

- Random Sampling: Every member has an equal chance of selection.

- Non-Random Sampling: Selection is based on non-probability methods like convenience or judgment.

###8. Define and give examples of qualitative and quantitative data.

- Qualitative Data: Descriptive (e.g., colors, names, labels)

- Quantitative Data: Numeric (e.g., height, weight, age)

###9. What are the different types of data in statistics?
-> The different types of data in statistics are:

- Nominal

- Ordinal

- Interval

- Ratio

###10. Explain nominal, ordinal, interval, and ratio levels of measurement.

- Nominal: Categories without order (e.g., gender)

- Ordinal: Categories with order (e.g., rankings)

- Interval: Numeric with order, no true zero (e.g., temperature)

- Ratio: Numeric with true zero (e.g., weight, height)

###11. What is the measure of central tendency?

-> It refers to statistical measures that identify the center of a dataset: mean, median, and mode.

###12. Define mean, median, and mode.

- Mean: The mean is the average of a set of numbers.
It is calculated by adding all the values and dividing by the total number of values.

- Median: The median is the middle value in a sorted dataset.If the number of values is odd, the median is the middle one.If even, it is the average of the two middle values.

- Mode:The mode is the value that appears most frequently in a dataset.


###13. What is the significance of the measure of central tendency?

-> The significance of the measure of central tendency is that it helps summarize large datasets and indicates where most values fall.

###14. What is variance, and how is it calculated?

->Variance measures data spread around the mean.It is calculated as:

       σ² = (1/n) * Σ (xi - μ)²

###15. What is standard deviation, and why is it important?

-> Standard deviation is the square root of variance. It shows how much data varies from the mean.

###16. Define and explain the term range in statistics.

-> In statistics, the range is a measure of spread that shows the difference between the highest and lowest values in a dataset.


###17. What is the difference between variance and standard deviation?

-> Variance is the average squared deviation while Standard Deviation is its square root, in the same units as the data.

###18. What is skewness in a dataset?
-> Skewness measures the asymmetry of the distribution:

- Positive skew: tail on the right

- Negative skew: tail on the left

###19. What does it mean if a dataset is positively or negatively skewed?

-> Positively Skewed: Mean > Median

- Negatively Skewed: Mean < Median

###20. Define and explain kurtosis.
-> Kurtosis describes the "tailedness" of the distribution:

- High kurtosis: more outliers

- Low kurtosis: fewer outliers

###21. What is the purpose of covariance?
-> Covariance shows how two variables vary together:

- Positive: variables increase together

- Negative: one increases as the other decreases

###22. What does correlation measure in statistics?

-> Correlation quantifies the strength and direction of the linear relationship between two variables.

###23. What is the difference between covariance and correlation?
-> Covariance is the General relationship units depend on data while Correlation is the Standardized (between -1 and 1), unitless.
